```python
import os
import requests
from dotenv import load_dotenv
load_dotenv()
```

```python
def requestGithub(endpoint):
    token = os.getenv("GITHUB_APIKEY")
    baseUrl = "https://api.github.com"
    url = baseUrl+endpoint
    headers = {
            "Authorization": f"token {token}"
        }
    res = requests.get(url,headers=headers)
    return res.json()
```

```python
master = requestGithub('/repos/ironhack-datalabs/scavenger/branches/master')
head_tree_sha = master['commit']['commit']['tree']['sha']
print(f"tree sha: {head_tree_sha}")
tree = requestGithub("/repos/ironhack-datalabs/scavenger/git/trees/0d82bf748275dbc1239a11e9157b198d88da7383?recursive=1")
tree
```

```python
paths= [i['path'] for i in tree['tree']]
print(paths)
path_sh = [(i.split('/')[-1],i.split('/')[0]) for i in paths if "scavengerhunt" in i]
print(path_sh)
```

```python
import base64
path_sh.sort()
encoded = []
for (t,f) in path_sh:
    word = requestGithub(f"/repos/ironhack-datalabs/scavenger/contents/{f}/{t}")
    encoded.append(word['content'])
encoded
decoded = [base64.b64decode(i) for i in encoded]
```

```python
decode = [str(i) for i in decoded]
decode = [i.replace("b'",'') for i in decode]
decode = [i.replace("\\n'",'') for i in decode]
decode = " ".join(decode)
decode
```
